Almost Right-angled Triangles I
---

Let us call an integer sided triangle with sides $a\leq b\leq c$ _barely acute_ if the sides satisfy $a^2+b^2 = c^2 + 1$.

How many barely acute triangles are there with perimeter $\leq 25 000 000$?

The perimeter is $a+b+c$. It seems like there's a nifty geometric way of thinking of this.
The law of cosines holds for sure.
$$ c^2 = a^2 + b^2 - 2ab\cos(C) $$
So
$$
1 = 2ab\cos(c)\qquad \longrightarrow \qquad \frac{1}{2ab} = \cos(C)
$$
and
$$
a+b+c \leq 25\times10^6
$$

Most of these triangles will have a very small $\cos C$, and so $C \approx \pi/2$.

```python
#drawing the plane

x = []
y = []
z = []
largest_perimeter =25000000
for a in range(0, int(largest_perimeter/2), 1000000):
    for b in range(0, largest_perimeter - a, 1000000):
        for c in range(0, largest_perimeter - a - b, 1000000):
            x.append(a)
            y.append(b)
            z.append(c)
            
x = np.outer(np.array(x), np.ones(len(x)))
y = np.outer(np.array(y), np.ones(len(y)))
z = largest_perimeter - x - y

ax = plt.figure().add_subplot(projection = '3d')
ax.plot_surface(x,y,z, edgecolor = 'royalblue', lw = 1)
ax.set(xlim = (0, largest_perimeter/2), ylim = (0,largest_perimeter), zlim=(0,largest_perimeter));
```

$a^2 + b^2 = c^2 + 1$ is an hyperbola. There is a right triangle with edges of length $a, b$ so that $a^2 + b^2 = x^2$. The equation becomes
$$
x^2 - c^2 = 1
$$
Again, c is always going to be pretty big. As big as perimeter$/2$. Because $a$ and $b$ have to be large for most of these integer edge triangles.



## approx equalateral --- "small" integers, a smallest, c smallest, b largest
Holding $C \approx \pi/2$ means it can be approx equilateral:
$$
c^2 \approx a^2 + b^2 \approx 2a^2\\
\longrightarrow c \approx \sqrt{2} a, a\approx b\\
\longrightarrow 25M \geq a+b+c \approx \sqrt{2}a + a + a\\
= \left(2+\sqrt{2}\right)a
\\
\longrightarrow a\leq 25M/(2+\sqrt{2})
$$
The side c will be smallest at this point, and so will the longer side of the triangle.
$$
a\leq 7 322 330. \dots \approx 7.32\times10^6\\
c\leq 10 355 339. \dots \approx 10.36\times 10^6
$$
with the equality for the perimeter at 25M and ignoring integer constraint.

## approx isosceles --- b smallest, a largest, c largest
In the case that $b->0$, $a^2 \approx c^2 + 1$. a will be half of 25M, and c will be half of 25M at most.
Making $a$ longer until $b->0$ makes
$$
c^2 \approx a^2 + b^2 \approx a^2\\
c \approx a \leq 12.5\times10^6
$$

There is an hyperbola in $x, c$ space with $x = \sqrt{a^2+b^2}$ When x is the sum of two squares and intersects the curve up to some distance, there may be a solution.

```python
ax = plt.figure()

x = np.linspace(0, 12.5e6, 1000)
c = np.sqrt(x**2 - 1)

plt.scatter(x,c)
plt.xlabel('x = sqrt(a**2 + b**2)')
plt.ylabel('c')
```

# actual squares

if $a^2 + b^2 = c^2 + 1$...

Draw a c-sided square and put an a-sided square in one corner. The leftover bars will have the area $b^2 - 1$.
If each bar is $c-a = n$ wide, then this area is
$$
n^2 + 2an = (c-a)^2 + 2a(c-a) = c^2 -2ac + a^2 + 2ac - 2a^2 =  c^2 - a^2
= b^2 -1
$$
Oh. This is just where we started.

We could scan through all as and differences $n$ and just... see if the result is one less than a square.
$a$ goes up to 12.5e6. The difference $n$ gets as large as $c-a = 10355339-7322330 = 3033009$
