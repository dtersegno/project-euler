Poker Hands
---

he card game poker, a hand consists of five cards and are ranked, from lowest to highest, in the following way:

High Card: Highest value card.
One Pair: Two cards of the same value.
Two Pairs: Two different pairs.
Three of a Kind: Three cards of the same value.
Straight: All cards are consecutive values.
Flush: All cards of the same suit.
Full House: Three of a kind and a pair.
Four of a Kind: Four cards of the same value.
Straight Flush: All cards are consecutive values of same suit.
Royal Flush: Ten, Jack, Queen, King, Ace, in same suit.
The cards are valued in the order:
2, 3, 4, 5, 6, 7, 8, 9, 10, Jack, Queen, King, Ace.

If two players have the same ranked hands then the rank made up of the highest value wins; for example, a pair of eights beats a pair of fives (see example 1 below). But if two ranks tie, for example, both players have a pair of queens, then highest cards in each hand are compared (see example 4 below); if the highest cards tie then the next highest cards are compared, and so on.

Consider the following five hands dealt to two players:

|Hand|	 	Player 1 hand	| P1 best |	Player 2	hand | p2 best| 	Winner| |
|---|---|---|---|---|---|---| 
|1|5H 5C 6S 7S KD|Pair of Fives|2C 3S 8S 8D TD|Pair of Eights|Player 2|
|2|5D 8C 9S JS AC|Highest card Ace|2C 5C 7D 8S QH|Highest card Queen|Player 1|
|3|2D 9C AS AH AC|Three Aces|3D 6D 7D TD QD|Flush with Diamonds|Player 2|
|4|4D 6S 9H QH QC|Pair of Queens, Highest card Nine|3D 6D 7H QD QS|Pair of Queens, Highest card Seven|Player 1|
|5|2H 2D 4C 4D 4S|Full House, With Three Fours|3C 3D 3S 9S 9D|Full House,with Three Threes|Player 1|


The file, poker.txt, contains one-thousand random hands dealt to two players. Each line of the file contains ten cards (separated by a single space): the first five are Player 1's cards and the last five are Player 2's cards. You can assume that all hands are valid (no invalid characters or repeated cards), each player's hand is in no specific order, and in each hand there is a clear winner.

How many hands does Player 1 win?

```python
'''High Card: Highest value card.
One Pair: Two cards of the same value.
Two Pairs: Two different pairs.
Three of a Kind: Three cards of the same value.
Straight: All cards are consecutive values.
Flush: All cards of the same suit.
Full House: Three of a kind and a pair.
Four of a Kind: Four cards of the same value.
Straight Flush: All cards are consecutive values of same suit.
Royal Flush: Ten, Jack, Queen, King, Ace, in same suit.'''

values = [str(value) for value in values]
value = dict(zip(values, range(len(values))))

suits = ['S','C','D','H']
royals = ['T','J','Q','K','A']
```

```python
import numpy as np

def get_cards(hand_string):
    the_cards = np.array([
        [card[0], card[1]] for card in hand_string.split()
    ])
    return the_cards

def is_flush(the_cards):
    cards = the_cards
    for suit in suits:
        if np.all(cards[:,1] == suit):
            return True
    return False

def all_royal(the_cards):
    #assume they are to start, change it to false if non royal encountered
    they_are_all_royal = True
    for card in the_cards:
        if card[0] not in royals:
            they_are_all_royal = False
    return they_are_all_royal

def is_royal_flush(the_cards):
    is_it_a_royal_flush = is_flush(the_cards) and all_royal(the_cards)
    return is_it_a_royal_flush, the_cards[0][1]

def two_match(the_cards):
    """
    Doesn't match two identical cards. That's cheating.
    """
    two_of_them = False
    for card1 in the_cards:
        for card2 in the_cards:
            if card1[0] == card2[0] and np.any(card1 != card2):
                two_of_them = True
                the_value = card1[0]
    return two_of_them, the_value

def three_match(the_cards):
    three_of_them = False
    for card1 in the_cards:
        for card2 in the_cards:
            for card3 in the_cards:
                if card1[0] == card2[0] == card3[0] and len(np.unique([card1[1],card2[1], card3[1]])) == 3:
                    three_of_them = True
                    the_value = card1[0]
    return three_of_them, the_value

def four_match(the_cards):
    four_of_them = False
    for card1 in the_cards:
        for card2 in the_cards:
            for card3 in the_cards:
                for card4 in the_cards:
                    if card1[0] == card2[0] == card3[0] == card4[0] and len(np.unique([card1[1],card2[1],card3[1],card4[1]])) == 4:
                        four_of_them = True
                        the_value = card1[0]
    return four_of_them, the_value

def is_full_house(the_cards):
    #it's a full house (3 of a kind and 2 of a kind) if three match and there are 2 and only 2 values.
    do_three_match = three_match(the_cards)
    all_values = np.unique(the_cards[:,0])
    two_and_only_two_values = len(all_values)==2
    the_suit = the_cards[0][1]
    return (do_three_match and two_and_only_two_values), the_suit

def is_two_pair(the_cards):
    #it's a two pair if the collection only has 3 values and doesn't have 3 of a kind
    do_three_match, the_value = three_match(the_cards)
    all_values = np.unique(the_cards[:,0])
    three_and_only_three_values = len(all_values) == 3
    return ((not do_three_match) and three_and_only_three_values), the_value

def is_three_kind(the_cards):
    do_three_match, the_value = three_match(the_cards)
    do_four_match = is_four_kind(the_cards)
    return (do_three_match and not do_four_match), the_value

def is_four_kind(the_cards):
    #it's four of a kind if there are only two unique values in the whole
    #set and four match
    do_four_match, the_value = four_match(the_cards)
    all_values = np.unique(the_cards[:,0])
    two_and_only_two_values = len(all_values) == 2
    return (do_four_match and two_and_only_two_values), the_value

def is_straight(the_cards):
    all_values = the_cards[:,0]
    number_values = np.sort([value[card_value] for card_value in all_values])
    # print(number_values)
    return np.all(np.diff(number_values) == 1)

def is_straight_flush(the_cards):
    return is_straight(the_cards) and is_flush(the_cards)

def highest_value_card(the_cards, vals = value):
    card_values = list(map(lambda x: x[0], the_cards))
    looking_for_card = True
    highest_value_card, highest_value = '2', value['2']
    for card_value in card_values:
        if value[card_value] > highest_value:
            highest_value_card, highest_value = card_value, value[card_value]
    return highest_value_card


def eval_hand(hand_string):
    these_cards = get_cards(hand_string)
    royal, suit = is_royal_flush(these_cards, 
    if royal:
        return 'Royal Flush', suit
    straight_flush,  = is_straight_flush(these_cards)
    if is_straight_flush(these_cards):####pick up here
        return 'Straight Flush'
    if is_four_kind(these_cards):
        return "Four of a Kind"
    if is_full_house(these_cards):
        return "Full House"
    if is_flush(these_cards):
        return "Flush"
    if is_straight(these_cards):
        return "Straight"
    if is_three_kind(these_cards):
        return "Three of a Kind"
    if is_two_pair(these_cards):
        return "Two Pair"
    if two_match(these_cards):
        return "One Pair"
    else:
        return highest_value_card(these_cards)

hands = [
'Royal Flush','Straight Flush',"Four of a Kind","Full House","Flush","Straight","Three of a Kind","Two Pair","One Pair"
] + values[::-1]

hand_value = dict(zip(hands[::-1], range(len(hands))))

def remove_first_value_card(the_cards, value_to_remove):
    removed_one = False
    new_cards = the_cards.copy()
    for card in new_cards:
        if card[0] == value_to_remove and not removed_one:
            card[0] = '2'
            removed_one = True
    return new_cards

def play_hands(player1, player2):
    """
    returns which player won the hand. Player 1 or player 2.
    
    DOESNT ACCOUNT FOR RANK OF PAIRS
    """
    hand1 = eval_hand(player1)
    hand2 = eval_hand(player2)
    val1 = hand_value[hand1]
    val2 = hand_value[hand2]
    if val1 > val2:
        return 1
    elif val2 > val1:
        return 2
    elif val1 == val2:
        p1_cards = get_cards(player1)
        p2_cards = get_cards(player2)
        highest1 = highest_value_card(p1_cards)
        highest2 = highest_value_card(p2_cards)
        new_p1_cards = p1_cards
        new_p2_cards = p2_cards
        while highest1 == highest2:
            new_p1_cards = remove_first_value_card(new_p1_cards, highest1)
            new_p2_cards = remove_first_value_card(new_p2_cards, highest2)
            highest1 = highest_value_card(new_p1_cards)
            highest2 = highest_value_card(new_p2_cards)
        if value[highest1] > value[highest2]:
            return 1
        elif value[highest2] > value[highest1]:
            return 2
        else:
            return 0
    else:
        return False
        
```
