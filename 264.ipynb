# Questions to answer:

- How to get triples of integer points with orthocenter at $H(0,x)$?

```python
import matplotlib.patches as patch
```

```python
fig, axes = plt.subplots(3,3)

for ax, point_set in zip(axes.reshape((-1,)), points):
    ax.plot(
        np.concatenate(
            [
                point_set[:,0],
                [point_set[0,0]]
            ]
        ),
        np.concatenate(
            [
                point_set[:,1],
                [point_set[0,1]]
            ]
        )
    )
    circle = patch.Circle((0,0), radius = np.linalg.norm(point_set[0]), fill = None, alpha = 0.3)
    ax.add_patch(circle)
    ax.grid()
    ax.set(xlim=(-10,10))
    ax.set(ylim=(-10,10))
```
