Triangle Containment
---

Three distinct points are plotted at random on a cartesian plane, for which
$
-1000\leq x, y \leq 1000$, such that a triangle is formed.

It can be verified that the triangle $ABC$ contains the origin:
$$
\begin{array}{rl}
A:& (-340,495)\\
B: &(-153,-910)\\
C: &(835,-947)\\
\end{array}
$$
whereas XYZ does not:
$$
\begin{array}{rl}
X: & (-175,41)\\
Y: &(-421, -714)\\
Z:& (574,-645)
\end{array}
$$

using `triangles.txt` find the number of triangles whose interior contains the origin.

---


Can show the origin is within the triangle by showing the line drawn from each vertex is within the angle created by the vertex.

```python
import numpy as np

def point_angle(px, py = None, deg = False):
    if type(px) == int or type(px) == float:
        x, y = px, py
    elif (len(px) != None) and py == None:
        x, y = px
    else:
        raise(ValueError('Give two values or a pair of values'))
    the_angle = np.angle(x + y*1j, deg=deg)
    return the_angle

def angle_between(point1, point2, deg = False):
    p1mag = np.linalg.norm(point1)
    p2mag = np.linalg.norm(point2)
    ap1p2 = np.math.acos( np.matmul(point1,point2)/(p1mag*p2mag))
    if deg:
        return ap1p2*180/np.pi
    else:
        return ap1p2
                         
def contains_origin(triangle):
    A, B, C = np.array(triangle)
    aAB, aBC, aCA = [
        angle_between(A,B),
        angle_between(B,C),
        angle_between(C,A)
    ]
    total = sum([aAB,aBC,aCA])
    small_float = sys.float_info.min
    does_it_contain_the_origin = total == 2*np.pi
    return does_it_contain_the_origin #, aAB, aBC, aCA, total
```

To be a triangle with the origin inside, the angle from the origin to any pair of points must add to 360 degrees.

Is the converse true? Probably. If the origin lies on a triangle edge, the angles are 90, 90, 180. Any further out and they all shrink.
