Spiral Primes
---

Starting with 1 and spiralling anticlockwise in the following way, a square spiral with side length 7 is formed.

```json
37 36 35 34 33 32 31
38 17 16 15 14 13 30
39 18  5  4  3 12 29
40 19  6  1  2 11 28
41 20  7  8  9 10 27
42 21 22 23 24 25 26
43 44 45 46 47 48 49
```

It is interesting to note that the odd squares lie along the bottom right diagonal, but what is more interesting is that 8 out of the 13 numbers lying along both diagonals are prime; that is, a ratio of 8/13 â‰ˆ 62%.

If one complete new layer is wrapped around the spiral above, a square spiral with side length 9 will be formed. If this process is continued, what is the side length of the square spiral for which the ratio of primes along both diagonals first falls below 10%?

---

First, a square with side of length n will encompass $n^2$ elements. The number of squares on the edge will be that number minus the size of the inside square. $n^2 - (n-2)^2 = 4n-4$. This is the length of four sides minus the number of overlapping corners.

The border of the square of width $n$ will start with the number $(n-2)^2 + 1 = n^2 -4n+5$ above the lower right corner.

````python
import numpy as np
import sympy as sp
import matplotlib.pyplot as plt
````

````python
def above_lower_right(length):
    return length**2-4*length+5
````

````python
list(map(above_lower_right, range(3,11,2)))
````

A square of width 3 will start with 2 above the lower right corner, width 5 will start with 10, 7:26, 9:50.
The upper right corner will be n-2 greater.

````python
def upper_right(width):
    return above_lower_right(width) + width - 2
````

````python
#list upper right diagonal, including center
list(map(upper_right, range(1,11,2)))
````

Upper left is $n-1$ greater.

````python
def upper_left(width):
    return upper_right(width) + width - 1
````

````python
list(map(upper_left, range(3,11,2)))
````

Lower left is another $n-1$ greater.

````python
def lower_left(width):
    return upper_left(width) + width - 1
````

````python
list(map(lower_left, range(3,11,2)))
````

And the lower right another.

````python
def lower_right(width):
    return lower_left(width) + width - 1
````

````python
list(map(lower_right, range(3,11,2)))
````

This confirms the odd squares are in the lower right diagonal.

````python
def upper_rights(n):
    return list(map(upper_right, range(3,n+1,2)))

def lower_rights(n):
    return list(map(lower_right, range(3,n+1,2)))

def upper_lefts(n):
    return list(map(upper_left, range(3,n+1,2)))

def lower_lefts(n):
    return list(map(lower_left, range(3,n+1,2)))
````

````python
#list all diagonal values.
def diagonal_values(n):
    if n%2 == 0:
        raise(ValueError('Use an odd width.'))
    else:
        diagonals = [1] + upper_rights(n) + upper_lefts(n) + lower_lefts(n) + lower_rights(n)
        # print(diagonals)
        diagonals = np.array(diagonals)
        return diagonals
    
def prime_diags(n):
    diavals = diagonal_values(n)
    l = len(diavals)
    num_primes = sum(map(sp.isprime, diavals))
    pct_primes = num_primes / l
    return [pct_primes, l]
````

````python
prime_diags(7)
````

````python
prime_diags(9)
````

````python
exes = np.arange(3,200,2)
whys = np.array(list(map(prime_diags, exes)))

plt.plot(exes, whys[:,0])
plt.title('fraction of primes on diagonals with odd square widths')
plt.xlabel('square width')
plt.ylabel('primes / total values');
````

````python
exes = np.arange(3,1000,2)
whys = np.array(list(map(prime_diags, exes)))

plt.plot(exes, whys[:,0])
plt.title('fraction of primes on diagonals with odd square widths')
plt.xlabel('square width')
plt.ylabel('primes / total values')
plt.axhline(0.1, c = 'blue')
````

It looks like it has to be a pretty big square! Let's look at wider spaced values to get a sense of where it may be.
